## Data exploration

```python
import xarray as xr
import xskillscore as xss
import glob
```

```python
# List train datasets
filenames = glob.glob("data/train_std/*")
```

```python
# Read train data
train_data = xr.concat([xr.open_zarr(filename) for filename in filenames], dim='time')
```

#### Histogram

```python
train_data.chirps.where(train_data.time.dt.month == 11, drop=True).mean(['longitude','latitude']).plot.hist(bins=20, density=True)
```

```python
train_data.chirps.where(train_data.time.dt.month == 12, drop=True).mean(['longitude','latitude']).plot.hist(bins=20, density=True)
```

```python
train_data.chirps.where(train_data.time.dt.month == 1, drop=True).mean(['longitude','latitude']).plot.hist(bins=20, density=True)
```

```python
train_data.chirps.where(train_data.time.dt.month == 2, drop=True).mean(['longitude','latitude']).plot.hist(bins=20, density=True)
```

```python
train_data.chirps.where(train_data.time.dt.month == 3, drop=True).mean(['longitude','latitude']).plot.hist(bins=20, density=True)
```

```python
train_data.chirps.where(train_data.time.dt.month == 4, drop=True).mean(['longitude','latitude']).plot.hist(bins=20, density=True)
```

```python
train_data.ecmwf.groupby(train_data.time.dt.strftime('%Y-%m')).mean(['longitude', 'latitude']).isel(time=slice(-100, -1)).plot.line()
train_data.chirps.groupby(train_data.time.dt.strftime('%Y-%m')).mean(['longitude', 'latitude']).isel(time=slice(-100, -1)).plot.line()
```

#### Correlation

```python
corr_time = xr.corr(train_data.chirps, train_data.ecmwf, dim=['longitude', 'latitude'])
```

```python
corr_time
```

```python
corr_space = xr.corr(train_data.chirps, train_data.ecmwf, dim=['time'])
```

```python
abs(corr_time).mean().load()
```

```python
abs(corr_space).mean().load()
```

```python
corr_time.groupby('time.month').mean().plot.line()
```

```python
corr_space.plot.imshow()
```

#### RMSE 

```python
rmse_time = xss.rmse(train_data.chirps, train_data.ecmwf, dim=['longitude', 'latitude'])
```

```python
train_data = train_data.chunk(dict(time=-1))
rmse_space = xss.rmse(train_data.chirps, train_data.ecmwf, dim=['time'])
```

```python
rmse_time.mean().load()
```

```python
rmse_space.mean().load()
```

```python
rmse_time.plot.line()
```

```python
rmse_space.plot.imshow()
```
